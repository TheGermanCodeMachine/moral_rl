### Contrastive rewards

```python
x = np.arange(100)
y_mcts = sorted(labels_mcts[:100])
y_step = sorted(labels_step[:100])
y_random = sorted(labels_random[:100])
plt.figure(figsize=(10,10))
plt.scatter(x, y_mcts, label='mcts')
plt.scatter(x, y_step, label='step')
plt.scatter(x, y_random, label='random')
plt.title('Contrastive Loss')
plt.legend()
plt.show()
```

```python
table = [
    ['measure', 'MCTS', 'Step', 'random'],
    ['mean', round(np.mean(y_mcts),2), round(np.mean(y_step),2), round(np.mean(y_random),2)],
    ['std', round(np.std(y_mcts), 2), round(np.std(y_step), 2), round(np.std(y_random), 2)],
    ['min', round(min(y_mcts),2), round(min(y_step),2), round(min(y_random),2)],
    ['max', round(max(y_mcts),2), round(max(y_step),2), round(max(y_random),2)],
]
print(tabulate(table, headers='firstrow'))
```

**MCTS**: 
- Has the smallest range (higher min, lower max), but is fairly evenly distributed within that range. 
    That might be because of the short trajectories which don’t have enough time to lead  to big difference in rewards. Notably this also explains the low score on validity.

**Step**: 
- The most negative one
- Has the highest variance, which naively should mean it’s easiest to predict (with low MSE)


**Random**:
- It’s much more positive than the others on average
- It has a stronger positive tail and less negative ones

**Overall**:
It is clear that the distributions are significantly different, both in their magnitude and variance. This will obviously have an impact on the learning of the model.

It’s also clear that the mean of mcts is closer to random than step is to random. We can assume that the models learn to predict roughly around the mean value of the labels in their training set. If that model is used on a different dataset with a different mean value in the labels it will still be biased to it’s own mean. This explains why mcts generalises better to random than step.
